# Libraries imported

```python
import torch
import torch.nn as nn
import torchaudio.transforms as TATransformers
from torch.utils.data import DataLoader, TensorDataset
```

```python
t=torch.randn(32, 1, 64, 64)
t=torch.flatten(t,start_dim=1)
print('Q1',t.shape[1])
```

```python
a = torch.randn(1, 16000)
mel_transform = TATransformers.MelSpectrogram(
    sample_rate=16000,
    n_fft=400,
    hop_length=160,
    n_mels=64
)
mt = mel_transform(a)
print('Q2',mt.shape)
```

```python
t=torch.randn(1, 1, 64, 64)
nn_conv = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1)
nn_maxpool = nn.MaxPool2d(kernel_size=2, stride=2)
t=nn_maxpool(nn_conv(t))
print('Q3',t.shape)
```

```python
nn_linear = nn.Linear(in_features=128, out_features=64)
sum=0
for p in nn_linear.parameters():
  if p.requires_grad:
    sum += p.numel()
print('Q4',sum)
```

```python
dt = TensorDataset(torch.zeros(1050))
dl = DataLoader(dt, batch_size=32, drop_last=False)
print('Q5',len(dl))
```

```python
t1 = torch.tensor([[2.5, 1.0, 0.1]])
t2 = torch.tensor([0])
cel = nn.CrossEntropyLoss()
loss = cel(t1, t2)
print('Q6',loss)
```

```python
t=torch.tensor([0.5])
wt = nn.Parameter(t)
opt = torch.optim.SGD([wt], lr=0.01)
wt.grad = torch.tensor([0.2])
opt.step()
print('Q7',wt)
```
